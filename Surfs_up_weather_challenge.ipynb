```python
from matplotlib import style
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import extract
import datetime
```

```python
#to create connection 
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
#Helps in mapping 
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
```

```python
Base.classes.keys()
```

```python
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session =Session(engine)
```

```python
session.query()
```

```python
#count to count number of station and adding .all() to get result as list.
session.query(func.count(Station.station)).all()
```

```python
session.query()
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
#staring the session
session.query()
#Extracting the data for month of dec 
Temp_results=session.query(Measurement.date,Measurement.station,Measurement.tobs).\
filter(extract('month',Measurement.date)==12).all()
#Extracting Temperature for month of Dec and getting year only to do group by 
temp_results_year=session.query((extract('year',Measurement.date)),Measurement.station,Measurement.tobs).\
filter(extract('month',Measurement.date)==12).all()
```

```python
#creating dataframe to save our query 
df = pd.DataFrame(temp_results_year, columns=['date','station','Temperature'])
df
```

```python
group_data=df.groupby(['station','date']).describe()
group_data
```

```python
session.query()
# Extracting the data for june
Temp_results=session.query(Measurement.date,Measurement.station,Measurement.tobs).\
filter(extract('month',Measurement.date)==6).all()
#Extracting year to group by 
temp_results_year=session.query((extract('year',Measurement.date)),Measurement.station,Measurement.tobs).\
filter(extract('month',Measurement.date)==6).all()
```

```python
#Changing the results to dataframe 
df = pd.DataFrame(temp_results_year, columns=['date','station','Temperature'])
df
```

```python
group_data=df.groupby(['station','date']).describe()
group_data
```
